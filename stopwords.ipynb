```python
from nltk.corpus import stopwords
```

```python
import nltk
nltk.download('stopwords')
```

```python
corpus = " my name is manan , I am from India and currently studying Btech in Thapar University . I actually enjoy solving some problems when I know the logic behind them."
```

```python
stopwords.words("english")
```

```python
from nltk.stem import PorterStemmer
stemmer = PorterStemmer()
sentences = nltk.sent_tokenize(corpus)
type(sentences)
```

```python
## apply stopwords 
for i in range(len(sentences)):
    word = nltk.word_tokenize(sentences[i])
    word = [stemmer.stem(word) for word in corpus if word not in set(stopwords.words("english"))]
    sentences[i]=' '.join(corpus)
```
