```python
wor = "manan kapoor is Indian and he is a coder , he likes to study machine learning and also is eating some bananas right now"
```

```python
import nltk
```

```python
 nltk.download('maxent_ne_chunker_tab')
```

```python
nltk.download('words')
```

```python
words = nltk.word_tokenize(wor)
tag = nltk.pos_tag(words)
```

```python
# Option 1: Print the tree structure directly (fast)
result = nltk.ne_chunk(tag)
print(result)

# Option 2: Print with binary=True to see named entities more clearly
print("\nWith binary=True:")
result_binary = nltk.ne_chunk(tag, binary=True)
print(result_binary)

# Option 3: Extract just the named entities
print("\nNamed Entities found:")
for chunk in result:
    if hasattr(chunk, 'label'):
        print(f"{chunk.label()}: {' '.join(c[0] for c in chunk)}")
```
